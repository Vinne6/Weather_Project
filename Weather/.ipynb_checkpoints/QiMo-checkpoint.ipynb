```python
import pymysql
import pandas as pd
import numpy as np
import re
```

数据库转存

```python
def data_out_sql(sql):
    #连接数据库
    conn = pymysql.connect(
        host="localhost",
        user="root",
        password="1111",
        database="data_weather",
        charset="utf8",
    )
    # 创建游标
    cursor = conn.cursor()
    # 执行sql语句
    cursor.execute(sql)
    # 调出数据
    data = cursor.fetchall()
    # cols为字段信息 例如(('factory_id', 253, None, 6, 6, 0, False), ('szDeviceId', 253, None, 30, 30, 0, False),('update_time', 7, None, 19, 19, 0, False))
    cols = cursor.description
    # 执行
    conn.commit()
    conn.close()
    # 将数据truple转换为DataFrame
    col = []
    for i in cols:
        col.append(i[0])
    data = list(map(list, data))
    data = pd.DataFrame(data,columns=col)

    return data
```

额外相关属性提取

```python
#查询某日某城市空气质量
def from_TimeAndId_find_air(city_time,city_id):
    city_id = str(city_id)
    try:
        air_data = data_out_sql("select * from air_index where riqi like '"+city_time+"'")
        city_air = air_data[city_id][0]
    except:
        city_air = 0
    finally:
        return city_air
```

```python
#查询城市id
def from_CityName_find_id(city_name):
    city_data = data_out_sql("select * from dim_region where shortname like '"+city_name+"'")
    city_id = city_data['id'][0]
    return city_id
```

```python
#查询城市经度
def from_CityName_find_lng(city_name):
    city_data = data_out_sql("select * from dim_region where shortname like '"+city_name+"'")
    city_lng = city_data['lng'][0]
    return city_lng
```

```python
#查询城市纬度
def from_CityName_find_lat(city_name):
    city_data = data_out_sql("select * from dim_region where shortname like '"+city_name+"'")
    city_lat = city_data['lat'][0]
    return city_lat
```

```python
#基于apply遍历的循环处理（效率极高，远优于for）

#直接调取数据库中信息，速度较快
def func_df_apply(item):
    item['city_id'] = from_CityName_find_id(item['city']) #填入城市id
    item['city_lng'] = from_CityName_find_lng(item['city']) #填入城市经度
    item['city_lat'] = from_CityName_find_lat(item['city']) #填入城市纬度
    item['city_air'] = from_TimeAndId_find_air(item['riqi'],item['city_id']) #填入空气质量
    #print(item)
    return item
```

语句清洗

关于天气语句的清洗，我们决定最大限度内保存数据中已有的天气情况。
经过分析数据内的语言，我们将天气划分为“晴”、“阴”、“霾”、“雾”、“阵雨”等21种基本天气。其中浮尘和扬沙天气由于数据量较小且性质较为接近因此划归为一类，整理得到20中基本天气分类。
因此将类似“中到大雨”、“大到暴雨”等转变类语句替换为“中雨 大雨”、“大雨 暴雨”。
同时将其中的连接词“到”、“转”、“~”替换为空格“ ”

```python
def func_df_apply_1(item):
    #替换
    item['weather'] = re.sub('小到中雨', "小雨 中雨", item['weather']) 
    item['weather'] = re.sub('中到大雨', "中雨 大雨", item['weather']) 
    item['weather'] = re.sub('大到暴雨', "大雨 暴雨", item['weather']) 
    item['weather'] = re.sub('小到中雪', "小雪 中雪", item['weather']) 
    item['weather'] = re.sub('中到大雪', "中雪 大雪", item['weather']) 
    item['weather'] = re.sub('到', " ", item['weather']) 
    item['weather'] = re.sub('转', " ", item['weather']) 
    item['weather'] = re.sub('~', " ", item['weather']) 
    #print(item)
    return item
```

```python
weather = weather.apply(func_df_apply_1, axis=1)
weather['weather'] #清洗后weather列如下
```

天气语言量化处理

```python
#新建表weather_1方便量化处理天气语言信息
weather_1 = weather[['weather']]
weather_1['w_0'] = '0'
weather_1['w_1'] = '0'
weather_1['w_2'] = '0'
weather_1['w_3'] = '0'
weather_1.head()
```

此处我们将单列天气数据分化至四列中，全天只有一种基本天气的情况，w_0、w_1、w_2、w_3四列为同一种天气。全天有两种基本天气的情况，w_0、w_1列为第一种天气，w_2、w_3列为第二种天气。全天有三种基本天气的情况，w_0列为第一种天气，w_1、w_2列为第二种天气，w_3列为第三种天气。全天有四种基本天气的情况，w_0、w_1、w_2、w_3四列分别对应每种天气。

```python
#遍历表将天气数据分化到w_0、w_1、w_2、w_3四列中
def func_df_apply_2(item):
    num = item['weather'].count(' ') #确定当天天气数量

    #仅一种天气时
    if(num == 0):
        item['w_0'] = item['weather']
        item['w_1'] = item['weather']
        item['w_2'] = item['weather']
        item['w_3'] = item['weather']
        
    #两种天气时
    if(num == 1):
        w1 = item['weather'].split(' ')
        item['w_0'] = w1[0]
        item['w_1'] = w1[0]
        item['w_2'] = w1[1]
        item['w_3'] = w1[1]
        
    #三种天气时
    if(num == 2):
        w1 = item['weather'].split(' ')
        item['w_0'] = w1[0]
        item['w_1'] = w1[1]
        item['w_2'] = w1[1]
        item['w_3'] = w1[2]
        
    #四种天气时
    if(num == 3):
        w1 = item['weather'].split(' ')
        item['w_0'] = w1[0]
        item['w_1'] = w1[1]
        item['w_2'] = w1[2]
        item['w_3'] = w1[3]
#     print(num)
    return item
```

```python
weather_1 = weather_1.apply(func_df_apply_2, axis=1)
weather_1
```
