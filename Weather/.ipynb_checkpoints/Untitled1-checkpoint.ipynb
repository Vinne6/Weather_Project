```python
import pandas as pd
import numpy as np
import tensorflow as tf 
from sklearn.metrics import explained_variance_score,mean_absolute_error,median_absolute_error
from sklearn.model_selection import train_test_split 
```

```python
C1 = pd.read_csv('weather.csv',encoding='gbk')
C1
```

```python
cit_sta = C1[['city']]
c = cit_sta.drop_duplicates(subset=['city'],keep='first')
c = c.reset_index(drop=True) 
c
```

```python
wea_sta = C1[['weather']]
a = wea_sta.drop_duplicates(subset=['weather'],keep='first')
a = a.reset_index(drop=True) 
a
```

```python
wind_sta = C1[['wind_direction']]
b = wind_sta.drop_duplicates(subset=['wind_direction'],keep='first')
b = b.reset_index(drop=True) 
b
```

```python
C1['weather'] = pd.factorize(C1['weather'])[0].astype(np.uint16)
C1
```

```python
C2 = C1['city']=='三亚'
C3 = C1
C3 = C3[C2]
C3
```

```python
C4 = C3[['temperature_h','temperature_l','weather','E','S','W','N','min_wind_power','max_wind_power']]
C4
```

```python
x = C4[[col for col in C4.columns if col != 'weather']]
y = C4['weather']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=23)  
```

```python
FEATURES = ['temperature_h','temperature_l','E','S','W','N','min_wind_power','max_wind_power']
LABEL = 'weather'
feature_cols = [tf.feature_column.numeric_column(k) for k in FEATURES]
```

```python
regressor = tf.estimator.DNNRegressor(feature_columns=feature_cols,hidden_units=[50, 50],model_dir='my_model')
```

```python
def my_input_fn(x, y, num_epochs=None, shuffle=True):
    return tf.estimator.inputs.pandas_input_fn(x=x, y=y,num_epochs=num_epochs,shuffle=shuffle)
```

```python
regressor.train(input_fn=my_input_fn(x_train,y_train,num_epochs=None, shuffle=True), steps=2000)
```

```python
predictions = regressor.predict(input_fn=my_input_fn(x_test,y=None,num_epochs=1, shuffle=False))  
```

```python
import itertools
```

```python
yuce = list(p["predictions"] for p in itertools.islice(predictions, 6))
print("Predictions: {}".format(str(yuce)))
```

```python
eval_result = regressor.evaluate(input_fn=my_input_fn(x_test,y_test))
print('\nTest set accuracy: {accuracy:0.3f}\n'.format(**eval_result))
```
